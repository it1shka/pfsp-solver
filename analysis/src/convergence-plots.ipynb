```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy as sp
```

## Convergence Plots for GA

```python
def plot_convergence_plot(path: str) -> None:
  df = pd.read_csv(path)
  logs = df['message']
  best_values = []
  worst_values = []
  avg_values = []
  for log in logs:
    best = int(log.split('best: ')[1].split(',')[0])
    worst = int(log.split('worst: ')[1].split(',')[0])
    avg = float(log.split('avg: ')[1].split(',')[0])
    best_values.append(best)
    worst_values.append(worst)
    avg_values.append(avg)
  plt.plot(best_values, label='best')
  plt.plot(worst_values, label='worst')
  plt.plot(avg_values, label='avg')
  plt.legend()
  plt.xlabel('iteration')
  plt.ylabel('value')
  plt.show()
```

```python
plot_convergence_plot('../_outputs/basic.csv')
```

```python
plot_convergence_plot('../_outputs/medium.csv')
```

```python
plot_convergence_plot('../_outputs/hard.csv')
```
